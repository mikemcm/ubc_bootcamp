```python
import novice

def make_thumbnail(filename, width):
    picture = novice.open(filename) # open picture
    new_height = int(picture.height * float(width)/picture.width)
    picture.size = (width,new_height)
    return picture
```

```python
tn = make_thumbnail("flower.png",159)
tn
```

```python
b = make_thumbnail('biking.png',400)
b
```
